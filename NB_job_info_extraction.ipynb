```python
# Create a function to cast a NNS to a NN
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

def plural_to_singular(word):
    """
    Convert a plural noun (NNS) to its singular form (NN).

    Args:
        word (str): The plural noun to convert.

    Returns:
        str: The singular form of the noun.
    """
    lemmatizer = WordNetLemmatizer()
    singular_form = lemmatizer.lemmatize(word, 'n')  # 'n' indicates a noun

    return singular_form

# Create a function to extract the job title
def extract_job_title(text):
    """
    Extract the job title from a job description.

    Args:
        text (str): The job description.

    Returns:
        str: The job title.
    """
    # Tokenize the text
    tokens = word_tokenize(text)

    # Get the part of speech tags of the text
    pos_tags = pos_tag(tokens)

    # Extract the entities from the part of speech tags
    entities = ne_chunk(pos_tags)

    # Create a list of job titles
    job_titles = []

    # Loop over the entities
    for entity in entities:
        # Check if the entity is a noun chunk
        if hasattr(entity, 'label') and entity.label() == 'ORGANIZATION':
            # Get the words in the entity and join them
            words = [word for word, tag in entity.leaves()]
            entity = ' '.join(words)

            # Cast the entity to singular
            entity = plural_to_singular(entity)

            # Append the entity to the job titles list
            job_titles.append(entity)

    # Return the job titles
    return job_titles

job_titles = extract_job_title(text)
print(job_titles)
```
